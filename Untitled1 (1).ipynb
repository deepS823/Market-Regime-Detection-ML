```python
import yfinance as yf
import pandas as pd
import ta
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Download data
nifty = yf.download("^NSEI", start="2015-01-01")
vix = yf.download("^INDIAVIX", start="2015-01-01")

# Flatten multi-level columns from yfinance output if they exist
if isinstance(nifty.columns, pd.MultiIndex):
    nifty.columns = nifty.columns.get_level_values(0)
if isinstance(vix.columns, pd.MultiIndex):
    vix.columns = vix.columns.get_level_values(0)

# Features
nifty['rsi'] = ta.momentum.RSIIndicator(nifty['Close'], window=14).rsi()
nifty['ma50'] = nifty['Close'].rolling(50).mean()
nifty['ma200'] = nifty['Close'].rolling(200).mean()

nifty['returns'] = nifty['Close'].pct_change()
nifty['volatility'] = nifty['returns'].rolling(20).std()

# FIXED MERGE
vix = vix[['Close']].rename(columns={'Close': 'vix'})
nifty = nifty.merge(vix, left_index=True, right_index=True, how='inner')

# Target
nifty['target'] = (nifty['volatility'] > nifty['volatility'].median()).astype(int)

# Clean data
nifty = nifty.dropna()

features = ['rsi', 'ma50', 'ma200', 'vix']

X = nifty[features]
y = nifty['target']

print("Shape:", X.shape, y.shape)

# Train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = XGBClassifier()
model.fit(X_train, y_train)

# Evaluate
preds = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, preds))

# Strategy
nifty['prediction'] = model.predict(X)
nifty['strategy_return'] = nifty['returns'] * (1 - nifty['prediction'])

nifty['market'] = (1 + nifty['returns']).cumprod()
nifty['strategy'] = (1 + nifty['strategy_return']).cumprod()

# Plot
plt.plot(nifty['market'], label="Market")
plt.plot(nifty['strategy'], label="Strategy")
plt.legend()
plt.show()
```

```python
import numpy as np

# Sharpe Ratio
sharpe = (nifty['strategy_return'].mean() / nifty['strategy_return'].std()) * np.sqrt(252)

# Max Drawdown
cum = nifty['strategy']
roll_max = cum.cummax()
drawdown = (cum - roll_max) / roll_max
max_drawdown = drawdown.min()

print("Sharpe Ratio:", sharpe)
print("Max Drawdown:", max_drawdown)
```

```python
# Improved strategy (earns small return during risk-off)
nifty['strategy_return'] = nifty['returns'] * (1 - nifty['prediction']) + (0.02/252) * nifty['prediction']

# Recalculate strategy curve
nifty['strategy'] = (1 + nifty['strategy_return']).cumprod()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(nifty['market'], label="Market")
plt.plot(nifty['strategy'], label="Improved Strategy")
plt.legend()
plt.title("Market vs Improved Strategy")
plt.show()
```

```python
def sharpe_fn(returns):
    return (returns.mean() / returns.std()) * np.sqrt(252)

print("Market Sharpe:", sharpe_fn(nifty['returns']))
print("Strategy Sharpe:", sharpe_fn(nifty['strategy_return']))
```
